```python
W=5
X=[]
Y=[]
folders=["./eje/eje1/","./eje/eje2/","./eje/eje3/"]
test_folder=[False,False,True]
low_filter=0.05
K_augmentation=2
```

```python
import tensorflow as tf 
from tensorflow import keras
from os import listdir
from os.path import isfile, join
import numpy as np

test_sample=[]
for idx,folder in enumerate(folders):
    DATA_CSV=folder+"act.csv.txt"
    with open(DATA_CSV, 'r') as f:
        
        back=tf.keras.preprocessing.image.load_img( folder+str(0)+"T.jpg", color_mode="rgb", target_size=None,interpolation='nearest')
        back = keras.preprocessing.image.img_to_array(back)
        back=tf.image.rgb_to_grayscale(back)
        back=np.array(back)

        lines = f.readlines()
        for line in lines:
            print(line)
            data=line.split(",")
            act=data[0]
            ini=int(data[1])
            end=int(data[2])
            
            print(act,"[",str(ini),"]",str(end))
            for i in range(ini,end,int(W/2)):
                print("i:",i)
                seq=[]
                for s in range(i-W+1,i+1):
                    print("\ts:",s)
                    image=tf.keras.preprocessing.image.load_img( folder+str(s)+"T.jpg", color_mode="rgb", target_size=None,interpolation='nearest')
                    image = keras.preprocessing.image.img_to_array(image)
                    image=tf.image.rgb_to_grayscale(image)

    
                    image=np.array(image)
                    image=np.subtract(image,back)
                    image=image/255.0
                    image[image < low_filter]=0
                    seq.append(image)
                seq=np.array(seq)
                X.append(seq)
                Y.append(int(act)-1)
                test_sample.append(test_folder[idx])
    
X = np.array(X)  
Y = np.array(Y)  
Y = keras.utils.to_categorical(Y)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image, ImageOps
import math
from scipy.ndimage import rotate
import random

i_img=10

fig, ax = plt.subplots(W,1,sharex =False ,sharey=True,
                             figsize=(15,20))


for i in range(0,W):
    xi=np.copy(X[i_img][i])
    mat = np.reshape(xi,(120, 160))
    im = Image.fromarray( np.uint8(mat*255) , 'L')
    ax[i].imshow(im)
        
        
plt.show()
```

```python


def rotate(origin, point, angle):
    """
    Rotate a point counterclockwise by a given angle around a given origin.

    The angle should be given in radians.
    """
    ox, oy = origin
    px, py = point

    qx = ox + math.cos(angle) * (px - ox) - math.sin(angle) * (py - oy)
    qy = oy + math.sin(angle) * (px - ox) + math.cos(angle) * (py - oy)
    return qx, qy

def augmentData(xis):
    
    angle=random.randint(-5,5)
    tx=random.randint(-20,20)
    ty=random.randint(-20,20)
    flip=random.choice([True, False])
    flipV=random.choice([False, False])
    scale=random.uniform(1,1.25)
    scale2=random.uniform(0.75,1)
    
    new_xis=[]
    for xi in xis:



        mat = np.reshape(xi,(120, 160))
        im = Image.fromarray( np.uint8(mat*255) , 'L')

        width, height = im.size   # Get dimensions

        new_width=int(width*scale)
        new_height=int(height*scale)
        im = im.resize((new_width,new_height))
        left = (new_width - width)/2
        top = (new_height - height)/2
        right = left+ width
        bottom = top+ height
        im = im.crop((left, top, right, bottom))

        new_width=int(width*scale2)
        new_height=int(height*scale2)
        im = im.resize((new_width,new_height))
        left2 = (width - new_width)/2
        top2 = (height - new_height)/2
        new_im = Image.new('L', (width, height))
        new_im.paste(im, (int(left2), int(top2)))
        im=new_im


        im=im.rotate(angle)
        im = im.transform(im.size, Image.AFFINE, (1, 0, -tx, 0, 1, -ty))
        if(flip):
            im = ImageOps.mirror(im)
        if(flipV):
            im = ImageOps.flip(im)

        theta = math.radians(-angle)
        im=np.reshape(np.array(im),(120, 160,1))/255
        new_xis.append(im)
        
        for idx, f in enumerate(xi):
            for idy, f in enumerate(xi[idx]):
                xi[idx][idy]=xi[idx][idy]*random.uniform(0.9,1.1)+random.uniform(0.0,0.1)
                if(xi[idx][idy]<low_filter):
                    xi[idx][idy]=0
                if(xi[idx][idy]>1):
                    xi[idx][idy]=1
    new_xis=np.array(new_xis)     
    return new_xis
```

```python
i_img=255

fig, ax = plt.subplots(1,W,sharex =False ,sharey=True,
                             figsize=(15,20))

xis=np.copy(X[i_img])
print(xis.shape)
xis=augmentData(xis)
print(xis.shape)
for i in range(0,W):
    idx=i
    xi=np.copy(xis[i])
    mat = np.reshape(xi,(120, 160))
    im = Image.fromarray( np.uint8(mat*255) , 'L')

    ax[idx].imshow(im)
        
        
plt.show()
```

```python
#from sklearn.model_selection import train_test_split
#x_train , x_val , y_train , y_val = train_test_split(X , Y ,test_size=0.2, random_state=0)
x_train=[]
y_train=[]
x_val=[]
y_val=[]
for i in range(0,len(X)):
    if test_sample[i]:
        x_val.append(X[i])
        y_val.append(Y[i])
    else:
        x_train.append(X[i])
        y_train.append(Y[i])
x_train = np.array(x_train)  
y_train = np.array(y_train) 
x_val = np.array(x_val)  
y_val = np.array(y_val) 
```

```python
if K_augmentation>0:
    X2,Y2=[],[]
    for idx, f in enumerate(x_train):
        print(idx,"/",len(x_train))
        for k in range(0,K_augmentation):
            xis=np.copy(x_train[idx])
            yi=np.copy(y_train[idx])
            xis=augmentData(xis)

            print(xis.shape)
            print(yi.shape)
            X2.append(xis)
            Y2.append(yi)
        xi=np.copy(x_train[idx])
        yi=np.copy(y_train[idx])
        X2.append(xi)
        Y2.append(yi)


    x_train = np.array(X2)        
    y_train = np.array(Y2) 
    #x_train , x_val2 , y_train , y_val2 = train_test_split(x_train , y_train ,test_size=0.01, random_state=0)
```

```python
from tensorflow import keras

from tensorflow.keras import layers
from keras.optimizers import Adam
from keras.layers import MaxPooling2D,Activation, Dense

from tensorflow import Tensor
from tensorflow.keras.layers import Input, Conv2D, ReLU, BatchNormalization,\
                                    Add, AveragePooling2D, Flatten, Dense,Dropout,LSTM
from tensorflow.keras.models import Model

model = keras.Sequential()

model.add(
    tf.keras.layers.TimeDistributed(
        Conv2D(32, (7, 7), padding='same', strides = 2),
        input_shape=(W, 120, 160, 1)
    ))
model.add(Activation('relu'))

model.add(tf.keras.layers.TimeDistributed(Conv2D(64, (5, 5), padding='same')))
model.add(Activation('relu'))

#model.add(tf.keras.layers.TimeDistributed(MaxPooling2D((2,2), data_format = 'channels_first')))

model.add(tf.keras.layers.TimeDistributed(Conv2D(128, (5, 5), padding='same', strides = 2)))
model.add(Activation('relu'))    

model.add(tf.keras.layers.TimeDistributed(Conv2D(128, (3, 3), padding='same')))
model.add(Activation('relu'))

#model.add(tf.keras.layers.TimeDistributed(MaxPooling2D((2,2), data_format = 'channels_first')))

model.add(tf.keras.layers.TimeDistributed(Conv2D(256, (3, 3), padding='same', strides = 2)))
model.add(Activation('relu'))

model.add(tf.keras.layers.TimeDistributed(Conv2D(256, (3, 3), padding='same')))
model.add(Activation('relu'))

#model.add(tf.keras.layers.TimeDistributed(Conv2D(256, (3, 3), padding='same', strides = 2)))
model.add(Activation('relu'))    

#model.add(tf.keras.layers.TimeDistributed(MaxPooling2D((2,2))))

model.add(tf.keras.layers.TimeDistributed(Conv2D(512, (3, 3), padding='same', strides = 2)))
model.add(Activation('relu'))    
#model.add(TimeDistributed(MaxPooling2D((2,2), data_format = 'channels_first', name='pool1')))    

#model.add(TimeDistributed(Conv2D(32, (1, 1), data_format = 'channels_first')))
#model.add(Activation('relu'))    

model.add(tf.keras.layers.TimeDistributed(Flatten()))

#model.add(TimeDistributed(Dense(512, name="first_dense" )))

#model.add(LSTM(num_classes, return_sequences=True))
model.add(LSTM(512 , return_sequences=True))
model.add(LSTM(512))
model.add(Dense(512))
model.add(Dense(5, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train,y_train ,
                    batch_size=64,
            validation_data=(x_val,y_val),
            epochs=10)
# Guardar el Modelo
model.save("my_model_Escena-eje.h5")
```
